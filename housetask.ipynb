```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d=pd.read_csv("USA Housing Dataset.csv")
```

```python
d.head()
```

```python
#droping the unnecessary columns such as date,street, city, statezip, country
d.drop(['date','country'],axis=1,inplace=True)
d.head()
```

```python
d.info()
```

```python
d.describe()
```

```python
#checking for null values/missing valuess
d.isnull().sum()
```

```python
# converting catagorical to numeric
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
d['statezip']= lb.fit_transform(d['statezip'])
d['street']=lb.fit_transform(d['street'])
d['city']=lb.fit_transform(d['city'])
d
```

```python
#to find the relation between the price and other features
d.corr()['price']
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(d.corr(),annot=True)
plt.show()
```

```python
d.columns
```

```python
x = d[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated', 'street', 'city',
       'statezip']]
```

```python
y=d['price']
```

```python
from sklearn.feature_selection import mutual_info_regression
score = mutual_info_regression(x, y)
score
```

```python
score1 = pd.DataFrame(score,columns=["scores"])
score1
```

```python
feature =pd.DataFrame(d.columns,columns=["features"])
feature
```

```python

new_feature =pd.concat([feature,score1],axis=1)
new_feature
```

```python
new_feature=new_feature.nlargest(9,columns="scores")
new_feature
```

```python
d=d.loc[:,new_feature["features"].values]
d.head(4)
```

```python
X=d[["city","bathrooms","street","condition","bedrooms","sqft_living","sqft_basement","sqft_above"]]
y=d["price"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
X_train[:3]
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
Xs_train = sc.fit_transform(X_train)
Xs_test = sc.fit_transform(X_test)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(Xs_train,y_train)
```

```python
lr.score(Xs_test,y_test)*100
```

```python
from sklearn.ensemble import RandomForestRegressor
random = RandomForestRegressor()
```

```python
random.fit(X_test,y_test)
```

```python
random.score(X_test,y_test)*100
```

```python
y_pred = random.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("R-squared:", r2_score(y_test, y_pred))
```
